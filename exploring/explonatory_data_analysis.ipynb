```python
numeric_cols = df.select_dtypes(include='number').columns
print(numeric_cols)
```

```python
# Coefficient of variation
# below 10%       low
# between 10%-30% mid
# above 30%       high
cv = ((df[numeric_cols].std() / df[numeric_cols].mean()) * 100).sort_values(ascending=False)
# skewness of features
# 0 perfectly symmetric
# >=1  higly right skewed
# <=-1 highly left skewed
skew = df[numeric_cols].skew().sort_values(ascending=False)

# Kurtosis of Features
# =0 Normal
# >0 Heavy tails -> more outliers
# <0 Light tails -> Fewer outliers 
kurt = df[numeric_cols].kurt().sort_values(ascending=False)

print("Coefficient of Variation:")
print(cv)
print("\nSkewness:")
print(skew)
print("\nKurtosis:")
print(kurt)
```
